```python
Endereco_entrada= "Rio Branco 23"
nome_rua = ""
Quebra_endereco = Endereco_entrada.split()
for i in range(len(Quebra_endereco)):
  if Quebra_endereco[i].isnumeric():
    numero_casa = Quebra_endereco[i]
  else:
    nome_rua = nome_rua + Quebra_endereco[i] + " "

if nome_rua[-1]== " ":
   nome_rua = nome_rua.rstrip(nome_rua[-1])
   if nome_rua[-1]== ",":
    nome_rua = nome_rua.rstrip(nome_rua[-1])
if numero_casa[-1]== " ":
   numero_casa= numero_casa.rstrip(numero_casa[-1])
       
Endereco_saida = [nome_rua , numero_casa]
print(Endereco_saida)      
```
